## Упражнение № 44

```python
df = pd.DataFrame(np.arange(20).reshape(-1, 5), columns=list('abcde'))
```

```python
def swap_cols(column_to_swap: str, swapper_column: str):
    cols = list(df.columns.values)
    index_to_swap = cols.index(column_to_swap)
    swapper_index = cols.index(swapper_column)
    temp = cols[swapper_index]
    cols[swapper_index] = cols[index_to_swap]
    cols[index_to_swap] = temp
    return df[cols]

print(swap_cols('a', 'c'))
print(df[list(df.columns.values)[::-1]])
```

```python
nearest_rows = []
nearest_distance = []

# итерируем по строкам
for i, row in df.iterrows():
    curr = row
    rest = df.drop(i) 
    
    # инициализируем словарь, чтобы хранить евклидово расстояние для текущей строки
    e_dists = {}  
    
    # итерируем по остальным строкам для текущей строки    
    for j, contestant in rest.iterrows():        
        # вычисляем евклидово расстояние и обновляем e_dists
        e_dists.update({j: round(np.linalg.norm(curr.values - contestant.values))})        
    
    # обновляем ближайшую строку для текущей строки и значение евклидово расстояния
    nearest_rows.append(max(e_dists, key=e_dists.get))
    nearest_distance.append(max(e_dists.values()))

df['nearest_row'] = nearest_rows
df['dist'] = nearest_distance
df
```
